```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/국토교통부_교량_20220627_정제.csv')
df2 = pd.read_csv('data/2022도로교량현황조서_정제.csv')
```

```python
df1.info()
```

```python
df2.info()
```

```python
# 중복데이터 확인

duplicates = df1.duplicated(keep=False)
df1[duplicates]     
```

```python
# 중복데이터 확인

duplicates = df2.duplicated(keep=False)
df2[duplicates]    
```

```python
# 중복데이터 제거

df2 = df2.drop_duplicates(keep='first')
```

```python
# 필요없는 컬럼 제거

df1 = df1.drop(['교량시작점위도', '교량시작점경도', '교량종료점위도', '교량종료점경도'], axis=1)
df2 = df2.drop(['준공년도'], axis=1)
```

```python
# 설계활하중이 미상인 데이터 제거

df1 = df1.drop(df1[df1['설계활하중'] == '미상'].index)
```

```python
# subset에 지정한 컬럼 값이 중복인 데이터 확인

duplicates1 = df1[df1.duplicated(subset=['교량명', '시도명', '시군구명', '교량연장', '교량폭', '상부구조형식', '설계활하중', '준공후경과년수'], keep=False)]
len(duplicates1)
```

```python
# subset에 지정한 컬럼 값이 중복인 데이터 모두 제거

df1 = df1.drop_duplicates(subset=['교량명', '시도명', '시군구명', '교량연장', '교량폭', '상부구조형식', '설계활하중', '준공후경과년수'], keep=False)
```

```python
# subset에 지정한 컬럼 값이 중복인 데이터 확인

duplicates2 = df2[df2.duplicated(subset=['시설명', '시도', '시군구', '총길이', '총폭', '상부구조', '설계하중', '준공후경과년수'], keep=False)]
len(duplicates2)
```

```python
# subset에 지정한 컬럼 값이 중복인 데이터 모두 제거

df2 = df2.drop_duplicates(subset=['시설명', '시도', '시군구', '총길이', '총폭', '상부구조', '설계하중', '준공후경과년수'], keep=False)
```

```python
df2 = df2.rename(columns={'시설명':'교량명',
                          '시도':'시도명',
                          '시군구':'시군구명',
                          '총길이':'교량연장',
                          '총폭':'교량폭',
                          '상부구조':'상부구조형식',
                          '설계하중':'설계활하중'})
```

```python
# df1과 df2를 df1을 기준으로 조인

df3 = pd.merge(df1, df2, on = ['교량명', '시도명', '시군구명', '교량연장', '교량폭', '상부구조형식', '설계활하중', '준공후경과년수'], how='left')
```

```python
df3 = df3.dropna(subset=['경간수', '최대경간장'])
```

```python
df3 = df3.reset_index(drop=True)
```
