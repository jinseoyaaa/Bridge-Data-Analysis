```python
import pandas as pd
import warnings
from datetime import datetime

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/국토교통부_교량_20220627.csv', encoding = 'cp949')
```

```python
col = ['교량명', '시설물종별등급구분', '시도명', '시군구명', '교량시작점위도', '교량시작점경도', '교량종료점위도', '교량종료점경도', '교량연장', '교량폭', '차로수', '상부구조형식', '교량준공연도', '최종안전점검결과', '설계활하중']
mydf = df[col]
```

```python
mydf = mydf[(mydf['시도명'] == '서울특별시') | (mydf['시도명'] == '경기도')]
```

```python
mydf.info()
```

```python
mydf.describe()
```

```python
mydf.loc[mydf['시설물종별등급구분'] == 99, '시설물종별등급구분'] = '기타'
```

```python
mydf[mydf['시설물종별등급구분'] == '기타'][:3]
```

```python
mydf.loc[mydf['교량명'] == '사암교', '교량폭'] = 10
```

```python
mydf[mydf['교량명'] == '사암교']
```

```python
width_0 = mydf[mydf['교량폭'] == 0].index
mydf = mydf.drop(width_0)
```

```python
mydf[mydf['교량폭'] == 0]
```

```python
year_99 = mydf[mydf['교량준공연도'] == 2099].index
mydf = mydf.drop(year_99)
```

```python
mydf[mydf['교량준공연도'] == 2099]
```

```python
mydf['설계활하중'] = mydf['설계활하중'].str.replace(pat = r'\([^)]*\)', repl = r'', regex = True)
```

```python
mydf['준공후경과년수'] = datetime.today().year - mydf.loc[:, '교량준공연도']
```

```python
mydf['준공후경과년수'][:3]
```

```python
mydf = mydf.drop('교량준공연도', axis = 1)
```

```python
mydf.reset_index(drop = True, inplace = True)
```
