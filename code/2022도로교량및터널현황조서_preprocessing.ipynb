```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_excel('data/2022도로교량및터널현황조서.xls', sheet_name='2022_교량')
```

```python
data.head()
```

```python
data.columns
```

```python
# 필요없는 컬럼 제거
trash = ['도로종류', '노선명', '읍면동', '리', '유효폭', '높이', '하부구조', '기관구분1', '기관구분2', '기관구분3']
data = data.drop(trash, axis=1)
data.columns
```

```python
unique_values = data['시도'].unique()
print(unique_values)
```

```python
# 서울특별시, 경기도 데이터만 추출
data = data[data['시도'].isin(['서울특별시', '경기도'])]
unique_values = data['시도'].unique()
print(unique_values)
data.head()
```

```python
data.isnull().sum()
```

```python
result = data.loc[data['총폭'].isnull()]
result
```

```python
result = data.loc[data['총폭'] == 0, ['시설명', '시도', '총길이', '총폭', '교통량', '상부구조', '준공년도']]
result
```

```python
# 사암교, 호장교2, 가목교 제거
data = data[~data["시설명"].isin(["사암교", "호장교2", "가목교"])]
result = data[data['시설명'].isin(['사암교', '호장교2', '가목교'])]
result
```

```python
# 가업교 제거
data = data.loc[~((data["준공년도"] == 2014) & (data["시설명"] == "가업교"))]
```

```python
result = data.loc[data['준공년도'] == 8888, ['시설명', '시도', '총길이', '총폭', '교통량', '준공년도', '상부구조']]
result
```

```python
data.loc[(data["시설명"] == "세월교") & (data["준공년도"] == 8888), "준공년도"] = 2010
data.loc[(data["시설명"] == "역북교") & (data["준공년도"] == 8888), "준공년도"] = 1995
data.loc[(data["시설명"] == "설매실1교") & (data["준공년도"] == 8888), "준공년도"] = 1980
data.loc[(data["시설명"] == "설매실2교") & (data["준공년도"] == 8888), "준공년도"] = 2018
```

```python
result = data.loc[data['시설명'].isin(['세월교', '역북교', '설매실1교', '설매실2교']), ['시설명', '시도', '총길이', '총폭', '설계하중','교통량', '상부구조', '준공년도']]
result
```

```python
data = data.loc[data["준공년도"] != 8888]
unique_values = data['준공년도'].unique()
print(unique_values)
```

```python
# 준공년도를 준공후경과년수로 변환
data['준공후경과년수'] = datetime.today().year - data.loc[:, '준공년도']
data.head()
```

```python
result = data.loc[data['경간수'] == 0, ['시설명', '시도', '총길이', '총폭', '교통량', '상부구조']]
result
```

```python
data = data.loc[data["경간수"] != 0]
```

```python
result = data.loc[data['경간수'] == 0, ['시설명', '시도', '총길이', '총폭', '교통량', '상부구조']]
result
```

```python
unique_values = data['설계하중'].unique()
print(unique_values)
```

```python
data.loc[data['설계하중'] == 'DB-24(성능개선)', '설계하중'] = 'DB-24'
```

```python
unique_values = data['설계하중'].unique()
print(unique_values)
```

```python
# 인덱스 재정렬
data = data.reset_index(drop=True)
data.head()
```
